### 🍓 Redaccion del problema:
Se requiere desarrollar una herramienta interactiva que permita a los usuarios calcular las propiedades fundamentales de un rectángulo (figura geométrica de cuatro lados con ángulos rectos). El programa debe ser capaz de recibir las dimensiones de la base y la altura para determinar cuánto espacio ocupa la figura (área) y cuánto mide su contorno (perímetro), presentando un menú claro para el usuario.
### 🍓 Formulas a utilizar
 * Área del Rectángulo
Fórmula: $A = b \cdot h$

* Perímetro del Rectángulo
Fórmula: $P = 2 \cdot (b + h)$

```python
# ==========================================
# 📐 CALCULADORA GEOMÉTRICA: EL RECTÁNGULO 📐}
# Elaborado por Angela Michelle Correa Lopez
# Fórmulas utilizadas:
# 🟧 Área: A = base * altura
# 📏 Perímetro: P = 2 * (base + altura)
# ==========================================

import math

def area_rectangulo():
    """Función para calcular y mostrar el área"""
    print("\n[  CALCULANDO EL ÁREA DEL RECTÁNGULO  ]")
    base = float(input("👉 Ingresa la medida de la base: "))
    altura = float(input("👉 Ingresa la medida de la altura: "))
    area = base * altura
    print(f"✨ Resultado: El área es {area:.2f} unidades²")

def perimetro_rectangulo():
    """Función para calcular y mostrar el perímetro"""
    print("\n[  CALCULANDO EL PERÍMETRO DEL RECTÁNGULO  ]")
    base = float(input("👉 Ingresa la medida de la base: "))
    altura = float(input("👉 Ingresa la medida de la altura: "))
    perimetro = 2 * (base + altura)
    print(f"✨ Resultado: El perímetro es {perimetro:.2f} unidades")

def menu():
    while True:
        print("\n" + "—"*40)
        print("       MENÚ DE CÁLCULO GEOMÉTRICO")
        print("—"*40)
        print("1. Calcular Área del Rectángulo")
        print("2. Calcular Perímetro del Rectángulo")
        print("3. Salir")
        print("—"*40)
        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            area_rectangulo()
        elif opcion == "2":
            perimetro_rectangulo()
        elif opcion == "3":
            print("\n👋 ¡Programa finalizado! Gracias por calcular. 🎓✨")
            break
        else:
            print("⚠️  Opción no reconocida. Por favor, ingresa 1, 2, 3.")

# Punto de inicio del programa
if __name__ == "__main__":
    menu()
print("Programa realizado por Angela Michelle Correa Lopez NL = 13")
```
